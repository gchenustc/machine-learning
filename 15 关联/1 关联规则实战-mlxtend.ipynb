```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
                    
```

## mlxtend的使用

```python
# 构建数据
data={'ID':[1,2,3,4,5,6],
'Onion':[1,0,0,1,1,1],
'Potato':[1,1,0,1,1,1],
'Burger':[1,1,0,0,1,1],
'Milk':[0,1,1,1,0,1],
'Beer':[0,0,1,0,1,0]}

data = pd.DataFrame(data)
data_use = data[["Onion","Potato","Burger","Milk","Beer"]]
```

```python
fren_items = apriori(data_use, min_support=0.5, use_colnames=True)
```

```python
association_rules(fren_items,metric="lift", min_threshold=1.2)
```

## One-hot 编码

```python
data = [[1,3,4],
[2,3,5],
[1,2,3,5],
[2,5]]

# 将number转换成str
for index,row in enumerate(data):
    data[index] = list(map(str,row))
    
data = pd.DataFrame({"index":range(1,len(data)+1),"data_":data})
```

```python
# list -> str
index = data.drop("index",1) # return copy
data.data_ = data.data_.str.join(",")
data
```

```python
data = data.data_.str.get_dummies(",")
data
```

```python
fren_items = apriori(data, min_support=0.5, use_colnames=True)
fren_items
```

```python
association_rules(fren_items, metric="lift", min_threshold=1)
```

## 电影题材实战

```python
data = pd.read_csv("datasets/ml-latest-small/movies.csv")
data
```

```python
pd.options.display.max_columns = 20
data_use = data.drop(["movieId","title"],1).genres.str.get_dummies()
data_use
```

```python
fren_items = apriori(data_use, use_colnames=True, min_support=0.08)
fren_items
```

```python
association_rules(fren_items, metric="lift", min_threshold=1)
```
