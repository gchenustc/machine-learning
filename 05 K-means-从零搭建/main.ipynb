```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from KMeansClass import KMeans
```

```python
from sklearn import datasets
iris = datasets.load_iris()
data = iris['data']
target = iris['target']
feature_names = iris['feature_names']
iris_types = iris['target_names']
feature_names
```

```python
iris.keys()
```

```python
features_selected_index = 2,3
x_axis, y_axis = feature_names[features_selected_index[0]], feature_names[features_selected_index[1]]
x_axis
```

```python
# 展示数据
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
for index,iris_type in enumerate(iris_types):
    plt.scatter(data[:,features_selected_index[0]][target==index], data[:,features_selected_index[1]][target==index], label=iris_type)
plt.legend()
plt.title("label_know")
plt.subplot(1, 2, 2)
plt.scatter(data[:,features_selected_index[0]],data[:,features_selected_index[1]])
plt.title("label_unknow")
plt.show()
```

```python
# 选定数据
X_train = data[:,features_selected_index]
y_train = target

# 实例化K-means
kmeans = KMeans(n_clustres=3, max_iters=100)
clusters_centers, y_ = kmeans.fit(X_train)
y_ = y_.flatten()

# 绘图
plt.figure(figsize=(12, 5))
    # 原数据
plt.subplot(1, 2, 1)
for index,iris_type in enumerate(iris_types):
    plt.scatter(data[:,features_selected_index[0]][target==index], data[:,features_selected_index[1]][target==index], label=iris_type)
plt.legend()
plt.title("label_know")
    # 训练结果
plt.subplot(1, 2, 2)
for index,iris_type in enumerate(iris_types):
    plt.scatter(data[:,features_selected_index[0]][y_==index], data[:,features_selected_index[1]][y_==index], label=iris_type)
    plt.scatter(clusters_centers[index][0], clusters_centers[index][1], s=100, marker ='s', label='cluster_center')
plt.title("label_trained")
plt.legend()
plt.show()
```
